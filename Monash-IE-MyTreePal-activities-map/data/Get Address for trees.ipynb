```python
import pandas as pd
import numpy as np
import requests
```

```python
# function to get request from api
def getAddress(lat,lon):
    url = 'https://nominatim.openstreetmap.org/reverse'
    params = {
        "format":"json",
        "lat":lat,
        "lon":lon  
    }
    response = requests.get(url,params=params)
    return response.json()
```
